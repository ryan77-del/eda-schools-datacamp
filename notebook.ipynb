Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
#Importando pandas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
#Importando documento
df = pd.read_csv('schools.csv')
# Quais escolas de NYC têm os melhores resultados em matemática?
filtro = df['average_math'] >= 640 # Filtro para escolas, 80% de 800 é igual 640
best_math_schools = df[filtro][['school_name','average_math']].sort_values('average_math',ascending=False)

best_math_schools

best_math_schools.sort_values('average_math').plot('school_name','average_math',kind='barh',figsize=(10,6),legend= False,align='edge',color='blue')
plt.tight_layout()
plt.xticks(rotation=50)
plt.xlabel('Média Matemática')
plt.ylabel('Escolas')
plt.title('Desempenho da Escolas')
plt.xlim(600, 760)
plt.xticks(np.arange(600,760,20))
plt.show()
```

```python
# Quais são as 10 escolas com melhor desempenho com base nas pontuações combinadas do SAT?  
df['total_SAT'] = df['average_math'] + df['average_reading'] + df['average_writing']
# Ordenação em ordem descrescente,limitar a 10 escolas
top_10_schools = df[['school_name','total_SAT']].sort_values('total_SAT',ascending = False).head(10)

top_10_schools


sns.barplot(data=top_10_schools,y='school_name',x='total_SAT',hue='school_name')
plt.show()
```
